```python
import pandas as pd

gu_name = ['종로구','중구','용산구','성동구','광진구']
col = ['시설이름', '도로명주소', '지번주소','운영기관명','전화번호','급식장소','급식대상','급식시간','급식요일','운영시작일자','운영종료일자','위도','경도','데이터기준일자','경찰서명','경찰서부서명','경찰서부서코드','시도경찰청명']
df = pd.read_csv('서울시무료급식소.csv', names=col)
df
```

```python
df[['시','구','동','상세주소']] = df['지번주소'].str.split(' ',n=3,expand=True)
df
```

```python
for i in range(len(df)):
    if df.loc[i, '구'] not in gu_name:
        df.drop(i, inplace=True)
df
```

```python
df = df[['구','동']]
data1 = df.groupby('동').count()
data1.columns = ['행정동 갯수']
data1
```

```python
df.replace(to_replace={'동':{'갈월동':'남영동','동자동':'남영동','성수동1가':'성수동1가1동','자양동':'자양3동'}}, inplace=True)
        
data2 = df.groupby('동').count()
data2.columns = ['무료급식소개수']
data2
```

```python
data2 = data2.reset_index()

data2.rename(columns={'동':'행정동'}, inplace=True)
data2
```

```python
data2.to_csv('data2_행정동별_급식소개수.csv')
```
