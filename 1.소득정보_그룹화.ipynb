```python
df_seoul = df[df['adstrd_nm'].str.contains("서울특별시")]
df_seoul
```

```python
df_seoul[['시','구','동']] = df_seoul['adstrd_nm'].str.split(' ',n=2,expand=True)
df_seoul
```

```python
income = df_seoul.groupby('구')['ave_income_amt'].mean()
income = income.sort_values(ascending=False)
income
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'NanumGothic'

plt.figure(figsize=(10,5))
plt.bar(income.index,income)
plt.xticks(rotation=45)
plt.show()
```
