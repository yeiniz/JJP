```python
import pandas as pd

df1 = pd.read_csv('./data/Group2.csv', index_col=0)
df1.drop(['저체중률', '독거노인수'], axis=1, inplace=True)
df1.head()
```

```python
df2 = pd.read_csv('./data/group2_score.csv')
df2.drop('자치구', axis=1, inplace=True)
df2.head()
```

```python
df = pd.merge(df1, df2, on='행정동')
df = df.sort_values('점수', ascending=False)
df.reset_index(drop=True, inplace=True)
df.rename(columns={'고령내국인전체':'고령자수'}, inplace=True)
df
```

```python
df.to_csv('Group2_final.csv', index=False)
```
