```python
import pandas as pd

df_gro2 = pd.read_csv('./data/gro2.csv', index_col=0)
df_count = pd.read_csv('./data/data2_행정동별_급식소개수.csv', index_col=0)
```

```python
df = pd.merge(df_gro2, df_count, how='left', on='행정동')
df
```

```python
add = pd.read_csv('./data/자치구행정동연결.csv')
add.drop('Unnamed: 0', axis=1, inplace=True)
add.head()
```

```python
df.insert(0,'자치구',0)
df['자치구'] = df['자치구'].astype('str')
df.head(1)
```

```python
for i in range(len(df)):
    for j in range(len(add)):
        if add.loc[j,'행정동'] == df.loc[i,'행정동']:
            df.loc[i,'자치구'] = add.loc[j,'자치구']
```

```python
# 결측값 0으로 대체
df.fillna(0, inplace=True)

# 전체 데이터프레임 확인
pd.set_option('display.max_rows', None)
df
```

```python
# 저체중률 불러오기

wei = pd.read_csv('./data/저체중률.csv', skiprows=16)
wei = wei[['사무직','3.8']]
wei.columns = ['자치구','저체중률']
wei.head()
```

```python
group2 = pd.merge(df, wei, how='left', on='자치구')
group2.head()
```

```python
group2.to_csv('group2.csv')
```
